## Creating and reshaping arrays

`np.arange` builds a flat range of values, `reshape` folds it into a new shape.

```python
a = np.arange(12).reshape(3, 4)
print(a)
```

```python
a[2:]
```

## Zeros and ones

```python
zeros = np.zeros((2, 3))
print(zeros)
```

```python
ones = np.ones((2, 3))
print(ones)
```

## Higher-dimensional reshape

Reshaping into 3 dimensions and slicing along the first axis.

```python
a = np.arange(24).reshape(4, 3, 2)
a[1, :, :]
```
