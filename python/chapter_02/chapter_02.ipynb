### Monty Hall

Three doors, one is a winner, the other two are losers. User picks a door. Of the two remaining doors, at least one will be a loser, a loser door is opened and shown to the user. Then the user is offered to switch doors between the one they have selected or the previous.

```python
episode_count = 1000
door_count = 3

rng = np.random.default_rng()
doors = np.arange(door_count)

total_wins_no_switch = 0
total_wins_switch = 0

for _ in range(episode_count):
    winner_door = rng.choice(doors)
    user_first_selected_door = rng.choice(doors)

    doors_that_could_be_shown = doors[
        (doors != user_first_selected_door) &
        (doors != winner_door)]
    shown_door = rng.choice(doors_that_could_be_shown)

    # no switch case
    if user_first_selected_door == winner_door:
        total_wins_no_switch += 1

    # switch case
    remaining_doors = doors[
        (doors != user_first_selected_door) &
        (doors != shown_door)]

    user_second_selected_door = rng.choice(remaining_doors)
    if user_second_selected_door == winner_door:
        total_wins_switch += 1

print("total wins no switch:     ", total_wins_no_switch)
print("total losses no switch:   ", episode_count - total_wins_no_switch)
print(f"No switch win %:           {100 * total_wins_no_switch / episode_count:.2f}%")
print("---")
print("total wins switch:        ", total_wins_switch)
print("total losses switch:      ", episode_count - total_wins_switch)
print(f"Switch win %:              {100 * total_wins_switch / episode_count:.2f}%")
```

```python
episode_count = 100000
number_of_coins = 3

# This array is used to track how often a number of coins land on heads when flipping x coins. For each episode, we flip x coins.
# For simplicity let's say we are flipping 1 coin. after the flip, summing the number of heads, we can either have 1 or 0 heads.
# that sum of coins that landed heads is the index of this head array. the value at that index is the total count of episodes
# that have ended in that state.
#
# If we had 3 coins, the possible states after flipping the coins could either be 0, 1, 2, or 3 heads, 4 possible states. hence
# the +1 to the number of coins.
heads = np.zeros(number_of_coins + 1)
for i in range(episode_count):
    flips = np.random.randint(0, 2, number_of_coins) # generate a random array of 0s and 1s of len number_of_coins
    h, _ = np.bincount(flips, minlength=2) # sum the elements in flips that equal 0, or heads
    heads[h] += 1 # add 1 to the index of the heads count

prob = heads / episode_count
print("Probabilities: %s" % np.array2string(prob))
```
