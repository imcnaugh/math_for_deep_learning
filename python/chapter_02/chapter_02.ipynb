## Monty Hall

3 doors, one is a winner, the other two are losers. User picks a door. of the two remaining doors, at least one will be a loser, a loser door is opened and shown to the user. then the user is offered to switch doors between the one they have selected, or the previous.

```python
door_count = 3

total_wins_no_switch = 0
total_wins_switch = 0

for i in range(1000):
    doors = list(range(door_count))
    winner_index = int(np.random.random() * door_count)
    winner = doors[winner_index]

    user_first_selected_index = int(np.random.random() * door_count)
    user_first_selected = doors[user_first_selected_index]

    if winner == user_first_selected:
        total_wins_no_switch += 1
    else:
        total_wins_switch += 1

print("total wins no switch: ", total_wins_no_switch)
print("total wins switch:    ", total_wins_switch)
```
